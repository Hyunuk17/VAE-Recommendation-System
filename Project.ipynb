```python
import numpy as np
import pandas as pd

import torch
import torch.nn as nn
import torch.nn.functional as F
```

```python
# 학습 조건 설정
BATCH_SIZE = 64
EPOCHS = 100
```

```python
from torch.utils.data import Dataset # 사용자 정의 데이터셋을 만들기 위함
from torch.utils.data import DataLoader # 데이터셋을 미니배치로 나누고 로딩

class CustomDataset(Dataset):

  # 데이터 정의
  def __init__(self, x_data): # 초기화
    self.x_data = x_data

  # 이 데이터 셋의 총 데이터 수
  def __len__(self):
    return len(self.x_data)

  # 어떠한 idx를 받았을 때 그에 맞는 데이터를 반환
  def __getitem__(self, idx):
      x = torch.FloatTensor(self.x_data[idx]) # 텐서로 반환
      return x
```

```python
# AutoEncoder 모델 설계
class Encoder(nn.Module): # nn 모듈을 상속받아 기본적인 기능을 사용가능
    def __init__(self):
        super(Encoder,self).__init__()
        self.fc1_1 = nn.Linear(1233, 256) # 완전연결 계층
        self.fc1_2 = nn.Linear(1233, 256) # 완전연결 계층
        self.relu = nn.ReLU() # 활성화 함수
                        
    def encode(self, x):
        mu = self.relu(self.fc1_1(x)) # 인코딩된 평균
        log_var = self.relu(self.fc1_2(x)) # 로그 분산
                
        return mu,log_var
    
    def reparametrize(self, mu, log_var): # 잠재 변수를 샘플링
        std = log_var.mul(0.5).exp_() # 표준편차 계산
        eps = torch.FloatTensor(std.size()).normal_().to(mu.device) # 정규분포를 따르는 샘플링
    
        return eps.mul(std).add_(mu) # 잠재 변수 z 반환 : Sample * 분산 + 평균
    
    def forward(self,x): # 순전파
        mu, log_var = self.encode(x)
        reparam = self.reparametrize(mu,log_var)
        
        return mu,log_var,reparam # 분포의 평균과 분산, Reparameterization Trick한 z값 반환
        
encoder = Encoder() 
```

```python
class Decoder(nn.Module):
    def __init__(self):
        super(Decoder,self).__init__()
        self.fc1 = nn.Linear(256, 1233) # 완전연결 계층
        self.simoid = nn.Sigmoid() # 활성화 함수 sigmoid
    
    def forward(self,x): # 순전파
        out = self.fc1(x) # z값으로 output 생성
        out = self.simoid(out) # 0~1 값으로 변환
        
        return out # New Data 반환
        
decoder = Decoder()
```

```python
reconstruction_function = nn.MSELoss(size_average=False) # 평균 제곱 오차

def loss_function(recon_x, x, mu, log_var): # 손실함수 정의
    MSE = reconstruction_function(recon_x, x) # 손실 계산

    KLD_element = mu.pow(2).add_(log_var.exp()).mul_(-1).add_(1).add_(log_var) # KL발산 최적화를 통해 기본 관계 및 패턴을 학습
    KLD = torch.sum(KLD_element).mul_(-0.5)

    return MSE + KLD # 총 손실 반환

parameters = list(encoder.parameters())+ list(decoder.parameters()) # 인코더와 디코더의 파라미터를 묶음
optimizer = torch.optim.Adam(parameters, lr=0.0005) # 옵티마이저, 학습률 지정
     
```

```python
def train(encoder, decoder, train_loader):
  encoder.train() # 학습 모드로 설정
  decoder.train()
  train_loss = 0 # 손실 초기화

  for feature in train_loader: # 순회
    feature = feature
    optimizer.zero_grad() # 옵티마이저 기울기 초기화
    mu,log_var,reparam = encoder(feature) # 인코더에 feature 전달, 평균, 분산, 잠재 변수 얻음
    output = decoder(reparam) # 잠재 변수로 output 얻음
    loss = loss_function(output, feature, mu, log_var) # 손실 구함
    loss.backward() # 역전파-기울기 구함
    optimizer.step() # 파라미터 업데이트
    
    train_loss += loss.item() # 학습 손실 업데이트

  train_loss /= len(train_loader) # 평균 손실
  return train_loss # 출력하기 위해 반환
```

```python
def evaluate(encoder, decoder, train_loader):
  encoder.eval() # 평가 모드로 설정
  decoder.eval() # 파라미터가 업데이트 되지 않도록 함
  result = [] # 결과 저장할 리스트
 
  with torch.no_grad(): # 기울기 계산 비활성화 - 평가 단계에서는 필요 없음

    for feature in train_loader: # train_loader 순회
      feature = feature 
      mu,log_var,reparam = encoder(feature) # encoder
      output = decoder(reparam) # decoder
      result.append(output.cpu().numpy()) # 데이터 결과 집합에 저장

  result = np.concatenate(result) # 결과 리스트 연결하여 하나의 배열로 변환
  return result
```

```python
train_dataset = CustomDataset(shopping_metrix) # 입력 데이터셋 얻기

train_loader = DataLoader( # train_loader 생성, 데이터를 batch_size로 나누어 공급
  train_dataset,
  batch_size = BATCH_SIZE, # 64
  shuffle = False, # 데이터 증강 없음
  drop_last = False)

for epoch in range(1, EPOCHS + 1): # epochs만큼 반복
  train_loss = train(encoder, decoder, train_loader) # 학습 손실 계산
  print(f"\n[EPOCH: {epoch}], \tTrain Loss: {train_loss:.4f}") # 현재 epochs 및 손실 출력
     
```

```python
# 평가
result = evaluate(encoder, decoder, train_loader)
result # 확률분포 형태
     
```
